## **1. Metallic Binding**
- **Metallic bonds** form due to valence electrons being shared over many atoms.
- **Free electrons** move through the **positive ion lattice**, providing cohesion.
- **Classical Model:** Treats electrons as **particles** colliding as they move.
- **Quantum Model:** Sometimes necessary to explain behavior.

---

## **2. Historical Use of Metal Nanoparticles**
- **290–350 AD:** Early metal nanoparticle applications.
- **800 AD:** Damascus steel blades contained nanoparticles.
- **13th–15th Century:** Luster pottery, using metal nanoparticles for color effects.
- **Notre Dame Stained Glass:** Light scattering by nanoparticles created vibrant colors.
- **Lycurgus Cup:** A Roman artifact whose color changes depending on light source.

---

## **3. Surface to Volume Ratio Effects**
- **Nanoparticles behave differently from bulk materials** because:
  - A significant fraction of atoms are **surface atoms**.
  - **Surface atoms have higher energy** and influence physical properties.
- **Thermal Properties:**
  - **Melting point decreases** as particle size decreases.
  - **Atoms become more mobile** (less tightly bound).
  - The effect is particularly strong for **nanoparticles < 20 nm**.

---

## **4. Example Problem: Surface-to-Volume Ratio**
### **Problem Statement**
Compare the surface-to-volume ratio of a **10 nm** gold nanoparticle to that of a **1 mm** gold particle.  

### **Solution**
For a spherical particle:


$$\text{Surface-to-Volume Ratio} = \frac{\text{Surface Area}}{\text{Volume}} = \frac{4\pi r^2}{\frac{4}{3} \pi r^3} = \frac{3}{r}$$

For a **10 nm** $ 10^{-8}$ m particle:


$$ \frac{3}{10^{-8} \text{ m}} = 3 \times 10^8 \text{ m}^{-1}$$

For a **1 mm** $10^{-3}$ m particle:


$$\frac{3}{10^{-3} \text{ m}} = 3 \times 10^3 \text{ m}^{-1}$$

### **Conclusion**
- The **10 nm particle has a surface-to-volume ratio** of 3 times $10^8$, which is **100,000 times greater** than the 1 mm particle.
- This explains why nanoparticles have **higher reactivity** and **melting point depression**.

## **5. Example Problem: Catalytic Activity**
### **Problem Statement**
A **bulk platinum (Pt) catalyst** has a surface area of **5 m²/g**, while **Pt nanoparticles** have a surface area of **200 m²/g**.  
If a catalytic reaction occurs at the surface, how much more efficient is the nanoparticle catalyst compared to the bulk catalyst?

### **Solution**
The **efficiency** of a catalyst is often proportional to its surface area:


$$\text{Efficiency Factor} = \frac{\text{Surface Area of Nanoparticles}}{\text{Surface Area of Bulk Material}}$$


$$= \frac{200 \text{ m}^2/\text{g}}{5 \text{ m}^2/\text{g}}$$


= 40


### **Conclusion**
- **Pt nanoparticles are 40 times more efficient** as a catalyst compared to bulk Pt.
- This is why **nanoparticles are widely used in catalysis**, e.g., in fuel cells.

## **6. Thomson Equation for Melting Point Depression**
The **Thomson equation** quantifies how nanoparticle size affects melting point:


$$T_m(D) = T_m^{bulk} \left( 1 - \frac{C}{D} \right)$$

where:
- $ T_m(D) $ = melting point of the nanoparticle,
- $ T_m^{bulk} $ = bulk melting point of the material,
- $ D $ = diameter of the nanoparticle,
- $ C $ = material-dependent constant.

### **Example Problem: Melting Point Depression**
#### **Problem Statement**
Gold has a **bulk melting point** of $ 1064^\circ C $. The constant  C  for gold is approximately 3.5  nm.  
Calculate the melting point of a **5 nm gold nanoparticle**.

#### **Solution**

$$T_m(D) = 1064 \left( 1 - \frac{3.5}{5} \right)$$


$$T_m(5) = 1064 \times 0.3 = 319.2^\circ C$$


#### **Conclusion**
- A **5 nm gold nanoparticle** melts at **319°C**, much lower than the bulk value.
- This is why **nanoparticles behave differently** in thermal applications.

## **7. Optical Properties of Gold Nanoparticles**
- The **color of gold nanoparticles** changes with particle size due to **surface plasmon resonance (SPR)**.
- **Surface Plasmon Resonance (SPR):**
  - Free electrons in the nanoparticle **oscillate** in response to light.
  - Different **sizes and shapes** influence absorption and scattering.

---

## **8. Example Problem: Rayleigh Scattering**
### **Problem Statement**
Using **Rayleigh's scattering law**, explain why the sky appears blue and sunsets appear red.

### **Solution**
Rayleigh scattering states that:


$$I \propto \lambda^{-4} D^6$$

- **Shorter wavelengths (blue, ~450 nm)** scatter **much more** than longer wavelengths (red, ~700 nm).
- During the day, sunlight travels a **shorter path** through the atmosphere → **blue light scatters more**, making the sky blue.
- At sunset, sunlight passes through **more atmosphere** → **blue light is mostly scattered away**, and **red/orange light remains**.

### **Conclusion**
- Rayleigh scattering explains **sky color changes** due to light scattering by atmospheric molecules.

# **Key Takeaways**
- **Surface-to-Volume Ratio:** Nanoparticles have a **much higher** surface area than bulk materials.
- **Catalytic Efficiency:** Nanoparticles have **higher reactivity** due to increased surface area.
- **Melting Point Depression:** Smaller nanoparticles melt at much **lower temperatures** than bulk.
- **Rayleigh Scattering:** Explains why **blue light scatters more**, making the sky appear blue.
