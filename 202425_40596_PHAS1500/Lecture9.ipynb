## **1. Hybridization and sp Orbitals**  

- **Hybridization** occurs when atomic orbitals mix to form new, equivalent hybrid orbitals.  
- **sp Hybridization:**  
  - The **2s orbital** mixes with just **one** of the **2p orbitals** to form **two equivalent sp orbitals**.
  - The two sp orbitals are **arranged linearly**, with the remaining **2p orbitals perpendicular** to them.

    

### **1.1 sp¹ Hybridized Carbon & Bonding in Ethyne (C₂H₂)**
- In ethyne (acetylene), **each carbon is sp hybridized**, leading to:
  1. Two **C-H σ bonds**:
     
     $$C(sp^1) - H(1s)$$
  2. One **σ bond** between the two carbons:
     
     $$C(sp^1) - C(sp^1)$$
  3. Two **π bonds** (overlapping p orbitals), creating a **triple bond (C≡C)**:
     
     $$C \equiv C$$

## **2. Conjugation and Delocalisation**  
- **Conjugation** occurs when **double and single bonds alternate**, allowing **p orbitals to overlap**, forming a **delocalized π electron cloud**.
- **Effects of Conjugation:**
  - Leads to **characteristic spectral absorptions** (UV-visible).
  - As conjugation increases, the energy difference between bonding and anti-bonding π orbitals **decreases**, causing **absorption in the visible range**.

## **3. Carbon, Nitrogen, and Oxygen**
- **Electron configurations:**
  - Carbon:  
    
    $$1s^2 2s^2 2p^2 $$
  - Nitrogen:  
    
    $$1s^2 2s^2 2p^3$$
    
  - Oxygen:  
    
    $$1s^2 2s^2 2p^4$$
- **Lone pairs:**  
  - Nitrogen and oxygen have **unbonded electron pairs** (lone pairs), which affect molecular shape.
  - **Lone pairs repel more strongly than bonded electrons**, causing deviations from perfect tetrahedral angles.
  - Example:
    - **Methane (CH₄):** Perfect tetrahedral.
    - **Ammonia (NH₃):** Slightly compressed due to lone pair repulsion.
    - **Water (H₂O):** More compressed, with a **bond angle of 104.4°**.

## **4. Hydrogen Bonding & Van der Waals Forces**
### **4.1 Hydrogen Bonds in Water**
- **Hydrogen bonds** occur due to **strong dipole-dipole interactions**.
- Water’s hydrogen bonds result in:
  - **High boiling point**.
  - **Ice having a lower density than water** due to an open hexagonal structure.

### **4.2 Van der Waals Interactions**
- **Van der Waals forces** arise from **fluctuating dipoles** that create weak **attractive interactions**.
- **Factors influencing strength:**
  - **Size of electron cloud** (larger atoms/molecules have stronger Van der Waals forces).
  - **Proximity of molecules**.

### **4.3 Lennard-Jones Potential (LJ Potential)**
- Describes **Van der Waals interactions** mathematically:
  - **Attractive term** (dominant at longer distances).
  - **Repulsive term** (dominant at short distances, due to the Pauli exclusion principle).
  - The **Lennard-Jones potential** equation:
    
    $$U(r) = 4\epsilon \left[ \left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^6 \right]$$
    where:
    - $\epsilon $ = depth of the potential well (interaction strength).
    - $ \sigma $ = atomic radius.
    - $ r $ = interatomic separation.

## **5. Carbon Allotropes**
- **Carbon exists in different structural forms** due to its ability to form multiple bonding types.

### **5.1 Graphite**
- **Structure:**  
  - **Layers of carbon atoms** arranged in **hexagonal sheets**.
  - **Each carbon is sp² hybridized**, forming **σ and π bonds**.
  - **Weak interlayer forces** allow easy sliding between layers.
- **Properties:**
  - **Good electrical conductor** (due to delocalized electrons).
  - **High thermal resistance**.
  - **Self-lubricating** (low friction).
- **Applications:**  
  - **Pencils**, **electrodes**, **brake drums**.

### **5.2 Diamond**
- **Structure:**
  - **Each carbon forms four σ bonds**, creating a **3D tetrahedral network**.
  - **Strong covalent bonds** make it extremely hard.
- **Properties:**
  - **Transparent** and **colorless**.
  - **Hardest naturally occurring material**.
  - **Electrical insulator** but **excellent heat conductor**.
- **Applications:**  
  - **Cutting tools**, **jewelry**, **drill tips**.
