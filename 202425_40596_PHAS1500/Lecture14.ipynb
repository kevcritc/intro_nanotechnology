

# **1. Semiconductor Materials and Quantum Dots**
## **1.1 Direct vs. Indirect Band Gap**
- **Direct Band Gap**: Electrons transition directly from the conduction band to the valence band, emitting a photon (efficient for optoelectronic applications).
- **Indirect Band Gap**: A phonon is required to facilitate the transition, making photon emission less efficient.

## **1.2 Semiconductor Elements and Their Properties**
Semiconductors are typically made from elements in **group IV** of the periodic table, as well as from compounds involving elements from **groups III-V and II-VI**. These materials have **moderate band gaps** and can be engineered for specific electronic and optical properties.

| Silicon (Si)        | IV        | 1.12             | Microelectronics, Solar Cells |
| Germanium (Ge)      | IV        | 0.66             | High-speed transistors, Infrared optics |
| Gallium Arsenide (GaAs) | III-V | 1.43             | High-frequency electronics, LEDs, Solar cells |

| Zinc Oxide (ZnO)    | II-VI     | >3.3             | UV LEDs, Transparent electronics |

## **1.4 Effective Mass of Charge Carriers**
- The **steepness of the E vs. k curve** determines the effective mass.
- The **effective mass** ($ m^* $) of a charge carrier is given by:
  $$
  m^* = \hbar^2 \left( \frac{d^2E}{dk^2} \right)^{-1}
  $$
  where:
  - $ \hbar $ is the reduced Planck's constant.
  - $ k $ is the wavevector, related to momentum by $ p = \hbar k $.
  - $ E $ is the energy of the charge carrier.

# **2. Quantum Dot Synthesis and Surface Chemistry**
## **2.1 Bottom-Up Synthesis Methods**
Quantum dots (QDs) are typically synthesized using **colloidal chemical methods**, ensuring control over size, shape, and surface properties. Key synthesis methods include:
- **Hot Injection Synthesis**: Rapid precursor injection into a hot solvent, allowing controlled nucleation and growth.
- **Hydrothermal and Solvothermal Synthesis**: High-temperature reactions in aqueous or organic solvents.
- **Microemulsion Techniques**: Surfactant-stabilized nanodroplets help control QD size.

## **2.2 Surface Chemistry of Quantum Dots**
- **Surface coatings** control QD stability, dispersibility, and functionalization.
- Functional groups influence **chemical reactivity and biocompatibility**:
  - **Carboxyl (-COOH)**
  - **Amine (-NH₂)**
  - **Hydroxyl (-OCH₂CH₂OH)**
  - **Biotin (-CH₃)**
  - **Thiols (-SH)**
- **Importance of Surface Coating**:
  - Prevents **agglomeration** (colloidal stability).
  - Enhances **biocompatibility** for medical applications.
  - Enables **functionalization for biosensing and imaging**.

# **3. Applications of Quantum Dots**
## **3.1 Bioimaging and Medical Applications**
- **Fluorescent Labeling**: QDs conjugated with biomolecules for cellular tracking.
- **Deep-Tissue Imaging**: Near-infrared (NIR) QDs enable deeper penetration into tissues.
- **Cancer Detection**: QDs are used to detect and highlight tumor cells.
- **Drug Delivery Tracking**: Monitoring the movement of therapeutic drugs within the body.

## **3.2 Quantum Dot-Based Sensors**
- **Glucose Detection**: QDs functionalized with glucose oxidase (GOx) for fluorescence-based glucose sensing.
- **pH and Ion Sensing**: QDs respond to environmental changes, allowing highly selective chemical detection.
- **Heavy Metal Detection**: Functionalized QDs detect toxic metals like lead and mercury.

## **4. Electroluminescent Quantum Dot Devices**
### **4.1 How Electroluminescence Works**
- Electroluminescence (EL) occurs when electrical energy is directly converted into light.
- **Charge injection and recombination in QDs result in light emission**.
- Emission color depends on **quantum confinement**, where smaller QDs emit shorter wavelengths.

### **4.2 Structure of a QD-LED**
- **Anode (ITO-coated glass)** → Injects holes.
- **Hole Transport Layer (HTL)** → Facilitates hole movement.
- **QD Emissive Layer** → Recombination of electrons and holes produces light.
- **Electron Transport Layer (ETL)** → Guides electrons into the QD layer.
- **Cathode (Metal layer)** → Injects electrons.

### **Problem 2: Semiconductor Conductivity**
A silicon semiconductor has an electron mobility of **1350 cm²/Vs** and a hole mobility of **480 cm²/Vs**. If the electron carrier density is **10¹⁶ cm⁻³** and the hole carrier density is **10¹⁵ cm⁻³**, find the conductivity.

**Solution:**
1. Convert units if needed (S/m, m²/Vs, etc.).
2. Use the conductivity equation:
   $$
   \sigma = q (n \mu_e + p \mu_h)
   $$
   $$
   \sigma = (1.602 \times 10^{-19}) [(10^{16} \times 1350) + (10^{15} \times 480)]
   $$
3. Solve for $ \sigma $:
   $$
   \sigma = (1.602 \times 10^{-19}) (1.35 \times 10^{19} + 4.8 \times 10^{17})
   $$
   $$
   \sigma = (1.602 \times 10^{-19}) (1.398 \times 10^{19})
   $$
   $$
   \sigma = 2.24 S/m
   $$
   **Final Answer:** $ \sigma = 2.24 S/m $
