

## 1. Referencing and Citations in Essays

### 1.1 In-Text Citations
- Use a **numerical system** in square brackets, e.g., [1].
- Place the number **before** the full stop.
- Example: *"Light scattering in films is well understood [1]."*

### 1.2 Creating a Reference List
- List **references in order of appearance**.
- Format:
  ```
  [Number] Author(s), Title, Journal/Book, Volume, Pages, Year.
  ```
- Example:
  ```
  [1] A. Smith, "Optical Properties of Thin Films," *J. Optics*, vol. 45, pp. 23-30, 2020.
  ```

### 1.3 Managing References
- **Reference management tools:** EndNote, Zotero, Mendeley.
- **In LaTeX:** Use BibTeX with a numbered style (e.g., IEEEtran).
- **In Word:** Use the built-in citation manager under the “References” tab.

### 1.4 Acceptable Reference Styles
- The preferred style is **numerical referencing (1-3 format)**:
  1. Journal articles: *Author(s), Journal, Year, Volume, Pages*.
  2. Books: *Author(s), Title, Edition, Year, Publisher, Pages*.
  3. Websites: *Author(s), Access Date*.
  4. Lecture notes: *Author(s), Year, Date*.

### 1.5 What Makes a Good Reference?
- A **primary source** of information.
- **Accurate and reputable**.
- **Accessible** to readers.
- **Relevant** to the topic.
- Contains **all necessary details** for verification.
- **Consistent formatting**.

## 2. Importance of Physics Scaling Laws in Nanotechnology

### 2.1 Why Size Matters
- **At nanoscale**, surface effects dominate over bulk properties.
- Example: Nanoparticles have **lower melting points** than bulk materials.

### 2.2 Scaling of Forces
- Gravity becomes negligible.
- **Electromagnetic and surface tension forces** dominate.
- Example: **Van der Waals forces** cause water droplets to stick to surfaces.

### 2.3 Quantum Effects
- Classical mechanics **fails** at the nanoscale.
- **Quantum mechanics** determines behavior.
- Example: **Quantum dots** have discrete electron energy levels.

### 2.4 Applications in Technology
- **Scaling laws** help in designing nanoscale devices.
- Examples:
  - Miniaturization in **microelectronics** (e.g., transistors).
  - **Nanomedicine** (e.g., drug delivery, sensors).

## 3. Scaling in Physics

There is a general method for solving scaling problems.  These basic steps can be applied generally and course if your maths is a strong you may see that you can take some short cuts!

1. Identify or Derive the Governing Equation  
    - Determine the fundamental equation that describes the system.
    

2. Identify Length-Scale Terms  
    - Locate terms in the equation that involve a length scale.  
    - If necessary, use material property equations (e.g., density, conductivity, capacitance) to introduce length-dependent terms.
    

3. Rearrange the Equation  
    - Express the equation in a form where the target property (e.g., speed, force, time) is the subject.

4. Convert to a Proportionality Expression  
    - Replace the **equals sign** with **"proportional to" (∝)**.  
    - Ensure the equation contains only the **target term** and **length scale terms (L's)**.  
    - Set all other dimensionless constants to unity.
    

5. Determine the Scaling Exponent  
    - Rewrite the expression in the form $L^n$.  
    - Identify the value of $n$ (e.g., 1, 2, -1/2) based on the proportionality relation.
    

6. Write Down Known Values for the Given Scale  
    - Example: If the system is a boat, assign values such as $L_1 = 2$ (length) and $v_1 = 10$ (speed).
    

7. Write Down Known Values for the Target Scale  
    - Example: If scaling the boat, assign $L_2 = \text{new length}$.
    

8. Set Up the Scaling Equation  

$$ \frac{v_1}{v_2} = \left( \frac{L_1}{L_2} \right)^n $$



9. Solve for the Unknown Term  
    - Rearrange the equation to express the unknown in terms of the known values:  

$$v_2 = v_1 \times \left( \frac{L_2}{L_1} \right)^n$$

10. Substitute Values and Compute the Answer  
    - Insert numerical values and calculate the final result.


### 3.1 Electric Field
- **Electric field strength** is inversely proportional to characteristic size $ L $.

### 3.2 Capacitance
- **Capacitance increases** with characteristic size $ L $.

### 3.3 Inductance
- **Inductance is proportional** to characteristic size $ L $.

### 3.4 Heat Transfer Scaling
- **Energy needed** to heat a material scales with mass $ L^3 $.
- **Heat flux** follows **Newton’s Cooling Law**.

### 3.5 Scaling in Optics

### 3.6 Biological Scaling Laws
Biological systems follow specific scaling laws that impact metabolism, strength, and lifespan as size changes.

#### 3.6.1 Metabolic Scaling (Kleiber’s Law)
- **Metabolic rate** (energy consumption per unit time) scales with **body mass $M$ as**:

  $ \text{Metabolic Rate} \propto M^{3/4} $
  $=\propto L^{9/4} $

- This means **larger animals consume energy more efficiently per unit mass**.
  - Example: **Elephants have a lower metabolic rate per gram of tissue than mice.**
  - Smaller animals require **higher food intake per gram of body weight**.
  

Figure 1. Plot of Metabolic Rate versus mass for a few example animals.

#### 3.6.2 Strength and Bone Scaling**
- **Strength** depends on cross-sectional area **(muscle/bone strength $ S \propto L^2 $ **, while **weight scales with volume $ W \propto L^3 $ **.
- **Larger animals are proportionally weaker**:
  - Example: **An ant can carry many times its body weight, but an elephant cannot.**
  - **Bone thickness must increase disproportionately for large animals to support their weight**.

#### 3.6.3 Heart Rate and Lifespan**
- **Heart rate scales as**:

  $ \text{Heart Rate} \propto L^{-1/4} $

Larger animals have slower heart rates and longer lifespans.
  - Example, Mice (500 bpm, 2-year lifespan) vs. Elephants (30 bpm, 60+ years).
  - Across species, most mammals have about **a billion heartbeats per lifetime**.


---


#### 3.6.4 Effects of Scaling on Strength
- **Smaller animals**: High strength-to-weight ratio (e.g., ants can carry many times their weight).
- **Larger animals**: Need **thicker muscles** (e.g., elephants).
- **Scaling humans up**:
  - If doubled in height:
    - Strength increases by 4 times (remember proportional to surface area).
    - Body mass increases by 8 times (why - proportional to mass).
    - The person would **struggle to stand**.

---

#### 3.6.5 **Scaling Summary:**
| Biological Factor | Scaling Law | Implication |
|------------------|------------|------------|
| **Metabolic Rate** | $ M^{3/4} $ | Larger animals are more energy-efficient. |
| **Strength** | $ L^2 $ | Larger animals are proportionally weaker. |
| **Heart Rate** | $ L^{-1/4} $ | Larger animals have slower heart rates & longer lifespans. |
| **Heat Loss** | $ L^2 $ vs. $ L^3 $ | Smaller animals lose heat faster due to surface area to mass |

This explains why **biological structures adapt in specific ways to size changes**, influencing evolution, survival, and behavior!

## 4. Moore’s Law

- **Number of transistors doubles every 2 years**.
- Transistor size must shrink to avoid **doubling the chip's surface area**.
- Every two years:
  - **Area halves** $ L^2 $.
  - **Characteristic length $ L $ decreases** by **$\sqrt{2}$**.

## 5. Writing a Technical Report

### 5.1 Style Guidelines
- Target audience: **Technical magazine**.
- **Use formal language**:
  - Avoid contractions (e.g., "don’t" → "do not").
  - Avoid clichés.
- **Be concise**: Remove unnecessary words.
- **Reference facts** that are not **common knowledge**.
- **Use subheadings** for clarity.
- **Low detail** in introduction/conclusion, **high detail** in core sections.
- **Avoid personal opinions/emotions**.

## 6. Nanophysics - The Atom

### 6.1 Discovery of Atomic Spectra
- **1800s:** Spectroscopic studies revealed **distinct emission lines** for different gases.
- Example: **Hydrogen, Helium, Sodium, Potassium**.



Figure 4.  The Hydrogen spectral lines in the visible spectrum.

### 6.2 The Rydberg Formula
- General formula for **hydrogen spectral lines**:

$\frac{1}{\lambda}=R\left(\frac{1}{n_1^2} -\frac{1}{n_2^2}\right)$

where, $n_1$ and $n_2$ are $>0$, $n_2>n_1$, and $R = 1.097776*10^7 \text{m}^{-1}$
 

### 6.3 Classical View of the Atom
- **Coulomb’s Law** governs attraction between **electrons and nucleus**.
- Circular motion of electrons leads to **centripetal force**.


Figure 5.  The Classical model of the atom with the forces added.

Coulomb's law is the starting point with the nuclear charge $Ze$ and charge of the electron $e$.
\begin{equation}
F=\frac{kZe^2}{r^2} \tag{1}
\end{equation} 

This force is equal the centripetal force
\begin{equation}
F=\frac{mv^2}{r} \tag{2}
\end{equation} 

where $v$ is the speed of the electron, $m$ is the mass, and $r$ is the distance from the centre of the atom.
Equating 1 and 2 and simplifying gives,

\begin{equation}
mv^2=\frac{kZe^2}{r} \tag{3}
\end{equation} 

Note that by multipling by $\frac{1}{2} we arrive at an expression for the kinetic energy (K).
\begin{equation}
\frac{1}{2}mv^2=\frac{kZe^2}{2r}\tag{4}
\end{equation}


### 6.4 Energy of the Electron
- **Potential Energy (PE)**: Work done moving the electron to distance  r .
\begin{equation}
U=\frac{-kZe^2}{r}\tag{5}
\end{equation} 

- **Kinetic Energy (KE)**: Motion of electron around nucleus.
Remember from Eq. 4 we have:
\begin{equation}
K=\frac{1}{2}mv^2=\frac{kZe^2}{2r}\tag{6}
\end{equation}
- **Total Energy (E)**:

\begin{equation}
E=U+K \tag{7}
\end{equation}

Sub in Eq. 5 and Eq. 6 into 7 to get the total energy, E

\begin{equation}
E=-\frac{1}{2}\frac{kZe^2}{r} \tag{8}
\end{equation}

  - Classical mechanics predicts **electron should spiral into nucleus**, which does not happen.
  - This problem led to the **development of quantum mechanics**.
