# Localized Surface Plasmon Resonance (LSPR) to Semiconductors

## 1. Introduction to Localized Surface Plasmon Resonance (LSPR)

### 1.1 Definition and Concept
Localized Surface Plasmon Resonance (LSPR) is the collective oscillation of conduction electrons in metallic nanoparticles excited by light. This phenomenon occurs when the frequency of incident photons matches the natural frequency of surface electrons oscillating against the restoring force of positive nuclei.

### 1.2 Mathematical Representation
The resonance condition is determined by:

$\varepsilon_{NP} = -2\varepsilon_m$

where:
- $\varepsilon_{NP}$ is the permittivity of the metal,
- $\varepsilon_m$ is the permittivity of the surrounding medium.

### 1.3 Factors Influencing LSPR
- **Nanoparticle Size and Shape:** Smaller particles have higher resonance frequencies. Elongated or anisotropic particles (e.g., nanorods) shift LSPR to longer wavelengths (red-shift).
- **Dielectric Environment:** Changes in the surrounding medium shift the resonance wavelength.
- **Composition of Nanoparticles:** Gold and silver are commonly used due to their strong plasmonic properties.

### 1.4 Shifting LSPR to the Near-Infrared (NIR) Region
The LSPR peak can be tuned into the near-infrared (NIR) region (700-1200 nm) by:
- **Increasing Particle Size:** Larger nanoparticles have a red-shifted LSPR peak.
- **Changing Shape:** Nanorods, nanoshells, and nanostars exhibit elongated structures that shift resonance towards NIR.
- **Using Core-Shell Structures:** Coating gold nanoparticles with a dielectric material can further shift the plasmonic resonance.

### 1.5 Importance for Photothermal Therapy
- **Biological Tissue Transparency:** NIR light penetrates biological tissues more deeply (several millimeters to centimeters) with minimal absorption by water and hemoglobin.
- **Efficient Heat Generation:** When NIR light excites the LSPR of gold nanoparticles, the absorbed energy is converted into heat, leading to localised temperature increases.
- **Selective Cancer Therapy:** Functionalised nanoparticles selectively bind to cancer cells, allowing for targeted hyperthermia that destroys cancer cells while sparing healthy tissue.
- **Minimal Invasiveness:** Enables non-invasive or minimally invasive treatments, reducing the need for surgery.

### 1.6 Applications
- **Biosensing:** Used in lateral flow assays (e.g., pregnancy tests using Au nanoparticles). 
  
#### How Lateral Flow Assays Work
Lateral flow assays (LFAs) are simple diagnostic devices used to detect the presence of a target analyte in a liquid sample. They operate based on capillary action, where the sample flows through a test strip consisting of:
  1. **Sample Pad:** Absorbs and distributes the sample evenly.
  2. **Conjugate Pad:** Contains gold nanoparticles conjugated with antibodies that bind to the target analyte.
  3. **Nitrocellulose Membrane:** Features two lines: a test line (with immobilized antibodies specific to the analyte) and a control line (ensuring the test is functioning properly).
  4. **Absorbent Pad:** Wicks away excess fluid.
  
#### Detection Mechanism
When the sample is applied, it flows through the strip. If the target analyte is present, it binds to the gold nanoparticle-antibody conjugates. This complex then moves along the membrane and binds to the test line, producing a visible color due to the plasmonic properties of the gold nanoparticles. A second control line captures unbound conjugates to validate the test.

#### Examples
- Pregnancy tests (detecting human chorionic gonadotropin, hCG)
- Infectious disease diagnostics
- Drug screening tests

- **Photothermal Therapy:** Gold nanoparticles can be used for targeted cancer therapy.
- **Catalysis and Energy Applications:** Plasmonic nanoparticles enhance catalytic reactions and are used in solar energy conversion.

## 2. Introduction to Semiconductors

### 2.1 Energy Bands in Solids
In an atom, electrons occupy discrete energy levels. However, when a large number of atoms bond to form a solid, their outer electron energy levels merge into continuous bands:
- **Valence Band (VB):** The highest energy band that is fully occupied at absolute zero.
- **Conduction Band (CB):** The next available band where electrons can move freely.
- **Band Gap (Eg):** The energy difference between the valence and conduction bands, given by:

$$E_g = E_c - E_v$$
where:
- $E_c$ is the conduction band minimum,
- $E_v$ is the valence band maximum.

#### Example: Silicon and 3p Levels
Silicon (Si) atoms have electron configurations ending in **3s² 3p²**. When many Si atoms bond to form a crystal lattice, their 3p orbitals interact, leading to the formation of **bonding and antibonding states**:
- As the atomic spacing decreases, the interaction between Si atoms increases, and discrete energy levels broaden into bands.
- Initially, when only a few atoms interact, molecular orbitals form from the individual 3p atomic orbitals.
- As more atoms are added, these orbitals split into a nearly continuous density of states.
- The **bonding states** form the valence band, while the **antibonding states** form the conduction band.
- The energy separation between these bands is the **band gap**, approximately **1.1 eV** for silicon.

#### Influence of Atomic Spacing on Density of States
- **Large Atomic Spacing:** The energy levels remain discrete with a small number of available states.
- **Reducing Atomic Spacing:** Overlapping wavefunctions cause increased band formation and the density of states increases.
- **Very Small Atomic Spacing:** Strong interactions cause broad bands with a nearly continuous range of energy levels.
- **At the Equilibrium Lattice Constant:** The energy bands stabilize, defining the semiconductor properties such as band gap and effective mass of carriers.

#### Classification of Materials by Band Gap
- **Conductors:** No band gap, as the conduction and valence bands overlap.
- **Semiconductors:** Small band gap (~0.1-3 eV) allowing thermal excitation of electrons.
- **Insulators:** Large band gap (>3 eV) preventing electron excitation at room temperature.


### 2.2 Semiconductor Materials
- **Intrinsic Semiconductors:** Pure materials like silicon (Si) and germanium (Ge), where thermal excitation generates electron-hole pairs.
- **Extrinsic Semiconductors:** Doped semiconductors with additional charge carriers.
  - **n-type:** Introduces extra electrons using elements like phosphorus.
  - **p-type:** Introduces holes using elements like boron.

### 2.3 Electrical Conductivity in Semiconductors
At absolute zero, semiconductors behave as insulators. However, at higher temperatures:
- Thermal energy, $E = \frac{3}{2} kT)$, excites a small number of electrons to the conduction band.
- These electrons contribute to electrical conduction along with holes moving in the valence band.

### 2.4 Density of Free Carriers
The intrinsic carrier concentration is given by:

$n_i = \sqrt{N_c N_v} e^{-E_g/2kT}$
where:
- $N_c$ and $N_v$ are the effective densities of states in the conduction and valence bands.

### 2.5 Applications of Semiconductors
- **Transistors and Microelectronics:** Basis of modern computing.
- **Solar Cells:** Convert sunlight into electricity.
- **LEDs and Lasers:** Utilize semiconductor band structures for photon emission.
